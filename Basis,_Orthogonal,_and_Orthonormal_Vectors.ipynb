```python
#buat vector i dengan elemen (1,0) dan vector j dengan elemen (0,1)
import numpy as np
i = np.array([1, 0])
j = np.array([0, 1])
print(i)
print(j)
```

```python
#buktikan bahwa vector i dan vector j itu orthogonal
dot_product = np.dot(i, j)
print(f"Hasil dot product i dan j :{dot_product}. jika hasilnya 0 maka i dan j adalah orthogonal")
```
